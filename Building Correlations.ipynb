```python
import pandas as pd
```

```python
test = pd.read_csv(r"C:\Users\user\Downloads\archive\test.csv")
train = pd.read_csv(r"C:\Users\user\Downloads\archive\train.csv")
```

Combine the two csv files across the index

```python
df = pd.concat([test,train])
```

```python
pd.set_option('display.max_columns', None)
```

```python
df[:2]
```

Convert each Distinct Value in a Categorical Column to a Binary Variable Column (Simply mentioning whether the Variable exists or not)

```python
dummies = pd.get_dummies(df)
```

Curious as to what the Scatter Plot looks like for Relationships with Satisfication (Binary) and other Variables (Categorical 1-5)

```python
import matplotlib.pyplot as plt
```

```python
for column in dummies.columns:
    plt.scatter(x=dummies[column], y=dummies.satisfaction_satisfied)
    plt.show()
```
